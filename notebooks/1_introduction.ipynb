### 1. Description of a simple ballistic problem

Newton’s Second Law:

$$
    \sum_i \vec{F_i} = m \vec{a}
$$

a) Projection onto the x-axis:

$$
    \sum_i F_{xi} = m a_{x} \Rightarrow
    m a_{x} = 0 \Rightarrow
    a_{x} = \frac{dv_x}{dt} = 0
$$

$$
    \int\limits_{v_0 \cdot cos\theta}^{v_x} dv_x = 0 \Rightarrow
    v_x - v_0 \cdot cos\theta = 0 \Rightarrow
    v_x(t) = v_0 \cdot cos\theta
$$

$$
    x = \int\limits_0^t v_x(t) dt = \int\limits_0^t v_0 \cdot cos\theta dt = v_0 \cdot cos\theta \cdot t
$$

$$
    x = v_0 \cdot cos\theta \cdot t \Rightarrow L = v_0 \cdot cos\theta \cdot T
$$

$$
    \color{red} {\mathbf{T = \frac{L}{v_0 \cdot cos\theta} \quad (1)}}
$$

b) Projection onto the y-axis:

$$
    \sum_i F_{yi} = m a_{y} \Rightarrow
    m a_{y} = -m g \Rightarrow
    a_{y} = \frac{dv_y}{dt} = -g
$$

$$
    \int\limits_{v_0 \cdot sin\theta}^{v_y} dv_y = \int\limits_0^t -g dt \Rightarrow
    v_y - v_0 \cdot sin\theta = -g t \Rightarrow
    v_y(t) = v_0 \cdot sin\theta - g t
$$

$$
    y = \int\limits_0^t v_y(t) dt = \int\limits_0^t (v_0 \cdot sin\theta - gt) dt 
      = v_0 \cdot sin\theta \cdot t - \frac{gt^2}{2}
$$

$$
    \color{red} {\mathbf{H = v_0 \cdot sin\theta \cdot T - \frac{gT^2}{2} \quad (2)}}
$$

Substitute $T$ from equation (1) into equation (2):

$$
    H = L \cdot tg\theta - \frac{g L^2}{2 v_0^2} \frac{1}{cos^2\theta} = L \cdot tg\theta - \frac{g L^2}{2 v_0^2} (tg^2\theta + 1)
$$

Let us denote:

$$
    p = tg\theta
$$

Then:

$$
    \frac{g L^2}{2 v_0^2} p^2 - Lp + \frac{g L^2}{2 v_0^2} + H = 0 \Rightarrow p^2 - \frac{2v_0^2}{gL} p + 1 + \frac{2v_0^2 H}{gL^2} = 0
$$

$$
    \color{red} {\mathbf{D = \frac{v_0^4}{g^2 L^2} - \frac{2 v_0^2 H}{g L^2} - 1 \ge 0 \quad (3)}}    
$$

$$
    \color{red} {\mathbf{v_0 \ge \sqrt{g(H + \sqrt{H^2 + L^2})} \quad (4)}}
$$

$$
    \color{red} {\mathbf{p_{1,2} = \frac{v_0^2}{g L} \pm \sqrt{D} \quad (5)}}
$$

From expression (3), it follows that $p_{1,2} \ge 0$ for all values of $v_0$.

$$
    \frac{1}{cos^2\theta} = 1 + tg^2\theta \Rightarrow \frac{1}{cos\theta} = \sqrt{1 + tg^2\theta} = \sqrt{1 + p^2}
$$

From equation (1):

$$
    \color{red} {\mathbf{T = \frac{L}{v_0} \sqrt{1 + p^2} \quad (6)}}
$$


### 2. Determining the time of flight and launch angle

$$
    g = 9.8 \; m/s^2
$$

$$
    H = 1 \; m
$$

$$
    L = 1 \; m
$$

From (4):

$$
    v_0^{min} = \sqrt{g(H + \sqrt{H^2 + L^2})}
$$

```python
from math import sqrt, pi, atan

g = 9.8

H = 1.
L = 1.

S = sqrt(H**2 + L**2)

v0 = sqrt(g*(H + S))

print(f'v0 = {v0}')
```

Check (3): 

$$
    D = \frac{v_0^4}{g^2 L^2} - \frac{2 v_0^2 H}{g L^2} - 1 \ge 0
$$

```python
D = v0**4/g**2/L**2 - 2*v0**2*H/g/L**2 - 1

print(f'D = {D}')
```

Calculate (5):

$$
    p_{1} = \frac{v_0^2}{g L} - \sqrt{D}
$$

$$
    p_{2} = \frac{v_0^2}{g L} + \sqrt{D}
$$

```python
p1 = v0**2/g/L - sqrt(D)
p2 = v0**2/g/L + sqrt(D)

print(f'p1 = {p1}, p2 = {p2}')
```

Calculate angles:

$$
    tg\theta_1 = p1 \Rightarrow \theta_1 = arctg(p_1)
$$

$$
    tg\theta_2 = p2 \Rightarrow \theta_2 = arctg(p_2)
$$

```python
theta1 = atan(p1)*180/pi
theta2 = atan(p2)*180/pi

print(f'theta1 = {theta1}, theta2 = {theta2}')
```

Find $T$ from (6):

$$
    T = \frac{L}{v_0} \sqrt{1 + p^2}
$$

```python
T1 = L*sqrt(1 + p1**2)/v0
T2 = L*sqrt(1 + p2**2)/v0

print(f'T1 = {T1}, T2 = {T2}')
```
